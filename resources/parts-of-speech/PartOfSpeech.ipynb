<font color='#e31934'> Part-of-speech Tagging<br>
<font color='#19cbe3' >NLTK Taggers<br>  

*   contains classes for part-of-speech tagging, or simply tagging.
*   Uses NLTK package

<font color='#048f43' >Tag <br>
A “tag” specifies some property of a token. <br><br>
<font color='#8734c2'>Functions 

*   punkt: utilized for sentence-tokenization
*   averaged_perceptron_tagger: used for tagging for parts of speech (POS)


```python
# import nltk package for breaking down into tokens.
import nltk
# download 'punkt' to tokenize our sentences.
nltk.download('punkt')
# download 'averaged_perception_tagger' for tagging with parts of their speeches.
nltk.download('averaged_perceptron_tagger')
```

```python
sentence = "The small boy ate the pancakes from the restaurant"
```

<font color='#00d8d6'>Tokenization

```python
# nltk.word_tokenize(): a function from 'punkt' class in nltk package
token = nltk.word_tokenize(sentence)
print(token)
```

<font color='#00d8d6'> POS Tagging

```python
tags = nltk.pos_tag(token)
print(tags)
```

<font color='#00d8d6'> POS Tags Summary

*   **DT**: determiner (e.g. 'a', 'the', 'an', etc) 
*  **JJ**: Adjectives
*  **NN**: Noun
*  **NNP**: Proper Nouns
*  **VBP**: non 3rd person singular present (habits and routines, general facts and truths, and thoughts and feelings). 
*  **NNS**: Common Noun Plural form
*  **IN**: Prepositions and Subordinating Conjunctions
